# set_hatch()

```python
labels = ['A', 'B', 'C']
values = [2, 5, 3]

bar = plt.bar(labels, values)
bar[0].set_hatch('/')
bar[1].set_hatch('o')
bar[2].set_hatch('-')

plt.show()
```

```python
labels = ['A', 'B', 'C']
values = [2, 5, 3]

bar = plt.bar(labels, values)

list1 = ['/', 'o', '-' ]

for indx,shape in enumerate(list1):
    bar[indx].set_hatch(shape)
    
plt.show()
```

```python
labels = ['A', 'B', 'C']
values = [2, 5, 3]

bar = plt.barh(labels, values)

list1 = ['/', 'o', '-' ]

for indx,shape in enumerate(list1):
    bar[indx].set_hatch(shape)
    
plt.show()
```

```python
labels = ['A', 'B', 'C']
values1 = [2, 5, 3]
values2 = [3, 1, 2]

plt.bar(labels, values1, label="values1", width=0.3)
plt.bar(labels, values2, label="values2", width=0.3)

list1 = ['/', 'o', '-' ]

for indx,shape in enumerate(list1):
    bar[indx].set_hatch(shape)
    
plt.legend()
plt.show()
```
